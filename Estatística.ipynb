```python
import numpy as np
```

```python
# producao diaria de 2 empregados hipoteticos
empregadaA =  [70, 71, 69, 70, 70]
empregadoB =  [60, 80, 70, 62, 83]
```

```python
# media de produ√ßao
print('Media da producao:',
      '\nEmpregado A :', np.mean(empregadaA),
      '\nEmpregado B :', np.mean(empregadoB))
```

```python
from matplotlib import pyplot as plt

plt.plot(empregadaA, color='g')
plt.plot(empregadoB, color='r')
plt.ylabel('Producao')
plt.xlabel('Tempo')
plt.legend(['Empregado A', 'Empregado B'], loc='best')

plt.xticks([])

plt.show()
```
