```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("dataset.xlsx")
```

```python
df.head(10)
```

```python
# Neste caso podemos dizer que a CORRELAÇÃO é neutra ou que não exiiste correlação entre as duas variaves
sns.lmplot(x= "Ano", y = "Cilindradas", data= df);
```

```python
#Correlaçao forte Positiva entre as duas variaves
# Na medida que a variavel X aumenta a Y tambem aumenta
sns.lmplot(x= "Ano", y = "ValorTabela", data= df);
```

# Nova seção

```python
# Correlação forte NEGATIVA
#Na medida que a variavel X aumenta a variavel Y diminui
sns.lmplot(x= "Consumo_Litro_Km", y = "Cilindradas", data= df);
```

```python
#avaliando as correlações entre todas as variaveis
plt.rcParams["figure.figsize"] = (18,8)
ax = sns.heatmap(df.corr(), annot= True)
```
